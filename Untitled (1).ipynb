```python
import pandas as pd
import numpy as np
df = pd.read_csv("student_scores.csv")
math_scores = pd.to_numeric(df["math_score"], errors="coerce").dropna().to_numpy()
print("Math Scores Array:")
print(math_scores)
mean_value = np.mean(math_scores)
median_value = np.median(math_scores)
max_value = np.max(math_scores)
min_value = np.min(math_scores)
print("\nMean:", mean_value)
print("Median:", median_value)
print("Maximum:", max_value)
print("Minimum:", min_value)
normalized_scores = (math_scores - min_value) / (max_value - min_value)
print("\nNormalized Scores:")
print(normalized_scores)
```
