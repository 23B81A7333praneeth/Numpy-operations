```python
import pandas as pd
df = pd.read_csv(r"")
df['average_score'] = (
    df['math_score'] +
    df['science_score']
) / 2
print("Dataset with Average Score:")
print(df.head())
top_students = df.sort_values(
    by='average_score',
    ascending=False
).head(5)
print("\nTop 5 Students:")
print(top_students)
correlation = df['attendance'].corr(df['average_score'])
print("\nCorrelation between Attendance and Average Score:")
print(correlation)
gender_avg = df.groupby('gender')['average_score'].mean()
print("\nAverage Marks by Gender:")
print(gender_avg)
```
